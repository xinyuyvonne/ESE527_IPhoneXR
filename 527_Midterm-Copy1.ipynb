```python
import pandas as pd
import numpy as np 
import seaborn as sns
from matplotlib import pyplot as plt
%matplotlib inline
# from brewer2mpl import qualitative

```

```python
df_heads = pd.read_csv("Desktop/vgsale.csv")
df_heads
```

```python
df_noMV = df_heads.dropna(axis=0,inplace = False)
df_noMV.describe()
```

```python
df_needed = df_heads.drop(["NA_Sales","EU_Sales","JP_Sales","Other_Sales","Global_Sales"],axis=1, inplace = False)
df_needed.describe()
```

```python
df_mE = df_needed.dropna(axis=0,inplace = False)
df_mE
```

```python
df_cS = df_mE.sort_values(by="Critic_Score",inplace = False, ascending = False)
```

```python
df_cS = df_mE.sort_values(by="Critic_Score",inplace = False, ascending = False)
df_cS.reset_index(drop=True,inplace=True)

sliced_cS = df_cS.loc[0:49, ["Name","Critic_Score"]]
sliced_cS
```

```python
df_cC = df_mE.sort_values(by="Critic_Count",inplace = False, ascending = False)
df_cC.reset_index(drop=True,inplace=True)

sliced_cC = df_cC.loc[0:49, ["Name","Critic_Count"]]
sliced_cC
```

```python
df_uS = df_mE.sort_values(by="User_Score",inplace = False, ascending = False)
df_uS.reset_index(drop=True,inplace=True)

sliced_uS = df_uS.loc[0:49, ["Name","User_Score"]]
sliced_uS
```

```python
df_uC = df_mE.sort_values(by="User_Count",inplace = False, ascending = False)
df_uC.reset_index(drop=True,inplace=True)

sliced_uC = df_uC.loc[0:49, ["Name","User_Count"]]
sliced_uC
```

```python
storeStringColumns = []

for name_Column, value_Column in df_noMV.iteritems():
    if type(value_Column[2]) == str:
        storeStringColumns.append(name_Column)

convertToNums = df_noMV.columns.difference(storeStringColumns)

df_onlyNums = df_noMV[convertToNums]
f, ax = plt.subplots(figsize=(15, 15))
plt.title('Pearson Correlation')

sns.heatmap(df_onlyNums.astype(float).corr(), linewidths=0.3, vmax=1.0,square=True,
            cmap="Greens", linecolor='k', annot=True)
```

```python
df_noMV['User_Score'] = pd.to_numeric(df_noMV['User_Score'])
sns.jointplot(x='Critic_Score',y='User_Score',data=df_noMV,
              kind='hex', cmap= 'Purples', size=13)
```

```python
sns.jointplot('Critic_Score','Critic_Count',data=df_noMV,
              kind='hex', cmap='Blues', size=13)
```

```python
group_by_genre = df_noMV.groupby(['Genre']).sum().loc[:, 'Critic_Score':'User_Count']

plt.figure(figsize=(15, 15))
sns.set(font_scale=1)
plt.subplot(211)
sns.heatmap(group_by_genre.loc[:, 'Critic_Score':'User_Count'],annot=True, fmt = '.1f')
plt.title("The total score and amount of reviewers that a type of game has gained")
```

```python
df_new = df_noMV[(df_noMV['Platform'] == 'Wii') | (df_noMV['Platform'] == 'PS3') | (df_noMV['Platform'] == 'X360')]
```

```python

sales_for_diff_ratings = df_new.groupby(['Rating','Platform']).Global_Sales.sum()
sales_for_diff_ratings.unstack().plot(kind='bar',stacked=True, colormap= 'Oranges', 
                           grid=False, figsize=(15,15))
plt.title('The sales of games with different ratings')
plt.ylabel('The Sales')
```
